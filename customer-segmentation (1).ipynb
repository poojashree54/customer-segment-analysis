```python
import sklearn
from sklearn.cluster import KMeans
```

```python
df = train_dummies.fillna(0)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
km = KMeans(init='k-means++', n_clusters=3, n_init=10).fit(df)
```

```python
km.labels_
```

```python
from scipy.spatial.distance import cdist
K=range(1,20)
```

```python
wss=[]
for k in K:
    km=KMeans(init='k-means++', n_clusters=k, n_init=10).fit(df)
    wss.append(sum(np.min(cdist(df,km.cluster_centers_, 'euclidean'),axis=1)) / df.shape[0])
```

```python
plt.plot(K,wss, 'bx')
plt.show()
```

```python
import sklearn.metrics as metrics
```

```python
labeles=KMeans(n_clusters=9,random_state=200).fit(df).labels_
```

```python
metrics.silhouette_score(df,labeles,metric="euclidean",sample_size=1000,random_state=200)
```
